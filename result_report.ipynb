```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
line = "183583859.334,140281345.899,95740576.8825,56961115.2938,56706676.4791,56478231.1822,17743279.3228,56233259.8872"
results = np.array([float(x) for x in line.split(',')])
index = np.array([i for i in range(2,10)])
```

```python
plt.plot(index,results)
plt.xlabel('K clusters')
plt.ylabel('Sum of Error')
plt.title('Elbow Method')
plt.grid(linewidth=1)
plt.show()
```
