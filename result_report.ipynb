```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
line = "183583859.334,275633474.327,101642300.218,66566756.8141,64106064.5898,26290198.4549,62103458.6954,61038346.6737"
results = np.array([float(x) for x in line.split(',')])
index = np.array([i for i in range(2,10)])
```

```python
plt.plot(index,results)
plt.xlabel('K clusters')
plt.ylabel('Sum of Error')
plt.title('Elbow Method')
plt.grid(linewidth=1)
plt.show()
```
